# **Introduction**
This document describes the encoding challenges encountered and solutions adopted during the TEI Lex-0 conversion of Clark Hall’s *A Concise Anglo-Saxon Dictionary* (CAS) and *Sweet’s The Student’s Dictionary of Anglo-Saxon* (SWEET). Its purpose is to ensure consistent encoding across both resources while accommodating their individual characteristics.


# **General entry structure and identifiers**

Each dictionary article is encoded as
 ```xml
<entry xml:id="..." type="mainEntry" xml:lang="ang">
  ```
Its `@xml:id` is formed from a dictionary-specific prefix and a normalized version of the lemma: `CAS2.` for *Clark Hall* and `SWT.` for *Sweet*. Thus, a main entry may receive an identifier such as `CAS2.bæc` or `SWT.cild`. The identifier is based on the complete normalized headword rather than on its abbreviated or typographically segmented printed form.

The main `<entry>` contains the lemma form, entry-level grammatical information, senses, and any nested lexical structures belonging to the article. Their order follows the order in which they appear in the source whenever this can be represented without distorting the TEI structure.

# **Sense hierarchy and sense identifiers**

In both *CAS* and *Sweet*, each `<sense>` is assigned a hierarchical `@xml:id` based on the dictionary prefix and the lemma, for example `CAS2.lemma.1` or `SWT.lemma.1`. The first number identifies the main sense division, while the following number identifies the individual sense nested within it. For this reason, even an entry containing only one meaning is encoded with an outer
 ```xml
 <sense xml:id="CAS2.lemma.1">
 ```
 and an inner
 ```xml
 <sense xml:id="CAS2.lemma.1.1">
  ```
Additional senses at the same level receive consecutive identifiers such as `.1.2`, `.1.3`, and so on.

In *Clark Hall*, sense divisions are marked only by colons or semicolons. Consequently, an entry normally has a single outer sense division, `.1`, containing the individual senses `.1.1`, `.1.2`, `.1.3`, etc. Higher-level lexical divisions are represented through separate homonym entries rather than additional major sense divisions.

In *Sweet*, `|` and `||` explicitly introduce new major sense divisions. The first division is assigned `.1`, with its subordinate senses numbered `.1.1`, `.1.2`, and so on. A following division introduced by `|` or `||` is assigned `.2`, with subordinate senses `.2.1`, `.2.2`, etc.

Where homonyms are present, the homonym number becomes part of the identifier before the sense hierarchy. Thus, the first outer sense under the first homonym may receive an identifier such as `CAS2.lemma_1.1` or `SWT.lemma_1.1`, while its first nested sense is identified as `CAS2.lemma_1.1.1` or `SWT.lemma_1.1.1`. Further senses are numbered consecutively at the relevant hierarchical level.

# **Dealing with complex variant forms**

`CAS` **wealh-moru**, -more (a^1) f., -mora m. _carrot, parsnip_.

Here the use of the `a^1` (`1` is superscript in the printed form) should mean that in the first syllable (that is `wealh`) `a` is found instead of `ea`.
The question was:  should the variant(s) `-more (a^1)` be interpreted as two forms: `wealh-more` and `walh-more`? Or does `(a^1)` mean that  `-more` only combines with `walh-`?  Our proposed solution: `(a^1)` in this and similar cases means that a variant form `-more `can only be possible with `walh-`, giving us two variants `wealh-moru` and `walh-more`.


```xml
<form type="lemma">
   <orth>
      <seg>wealh</seg>
      <seg>-moru,</seg>
   </orth>
   <form type="variant">
      <orth extent="part" expand="walh-more">
      <seg>-more</seg> <seg>(a<lbl rend="sup">1</lbl>)</seg></orth>
   </form>
   ...
</form>
...
```

# **Dealing with complex inflected forms**

`CAS` **±blēo** n. gs. blēos, ds. blēo, gp. blēo(na), dip. blēom, blēo(w)um
_colour_, _Bo_; CP: _appearance, form_, _Sol_. [‘_blee_’]

Inflected forms are encoded with `<form type="inflected">`. The grammatical information applying to the form is recorded in `<gramGrp>`, while the form itself is represented in `<orth>`.
Where parentheses indicate two possible realizations of the same printed form, both expanded forms are supplied in `@expand`. Thus, `blēo(na)` is encoded as
```xml
<form type="inflected">
   <gramGrp>
      <gram type="case" value="genitive">g</gram>
      <gram type="number" value="plural">p.</gram>
   </gramGrp>
   <orth expand="blēo, blēona">blēo(na)</orth>
</form>

n some cases, two or more separate forms are given for the same inflectional category, as with `blēom, blēo(w)um` under `dip.`. Under the current policy, these are placed within a single `<form type="inflected">` as two separate `<orth>` elements sharing the same `<gramGrp>`. This solution is used for simplicity and preserves their association with the same grammatical specification
```xml
<form type="inflected">
   <gramGrp>
      <gram type="case" value="dative, instrumental">di</gram>
      <gram type="number" value="plural">p.</gram>
   </gramGrp>
   <orth>blēom</orth>
   <orth expand="blēoum, blēowum">blēo(w)um</orth>
</form>

This is not necessarily the most appropriate final model, however. A representation more closely aligned with TEI Lex-0 would encode each realization in a separate `<form type="inflected">`. Also, where several inflected forms belong to the same inflectional set, they could additionally be grouped within `<form type="paradigm">`. This alternative is noted for possible future revision but is not part of the current encoding policy.

**______________________________________________________________________________**

`SWEET` **wrēon** 6, 7 *prt.* **wrāh**, **ēa**, *pl.* **wrigon**, **u**, *ptc.* **wrigen**, **o** cover *rare in W*.

`SWEET` **cealf**, *A.* **ę**, **ǣ**, *n.* *pl.* **~as**, **~ru**, *A.* **calfur**, **calferu** calf.

Another alternative (currently implemented only for *Sweet* encodings) is treating additional inflected forms as orthographic variants nested in said form as in `prt. wrāh, ēa`, `pl. wrigon, u`, `ptc. wrigen, o`:
```xml
<form type="inflected">
   <gramGrp>
      <gram type="tense" value="past">prt.</gram>
   </gramGrp>
   <orth>wrāh</orth>
   <pc>,</pc>
   <form type="variant">
      <orth extent="part" expand="wrēah"><seg>ēa</seg></orth>
   </form>
</form>  

Or a little more complex with three levels of nested variants as in `pl. ~as, ~ru, A. calfur, calferu`:
```xml
<form type="inflected">
   <gramGrp>
      <gram type="number" value="plural">pl.</gram>
   </gramGrp>
   <orth extent="part" expand="cealfas">
      <seg>~as</seg>
   </orth>
   <pc>,</pc>
   <form type="variant">
      <orth extent="part" expand="cealfru">
         <seg>~ru</seg>
      </orth>
   </form>
   <pc>,</pc>
   <form type="variant">
      <usg type="geographic" norm="Anglian">A.</usg>
      <orth>calfur</orth>
      <pc>,</pc>
      <form type="variant">
         <orth>calferu</orth>
      </form>
   </form>
</form>

**______________________________________________________________________________**

`SWEET` **habban**, **æ**, (*hē*) **hæfþ**, **hafaþ**, *prt.* **hæfde** *w. a., partitive g.* have, hold, take, *met.* possess, hold, have: **~ wiþ þām fȳre**; **~ mid him**; **æt hæbbendre handa** with the stolen goods in his possession; **wīf ~** have intercourse with; **hē hæfde þrītig wintra** he was thirty; **gif hē þæs andgytes hæfþ þæt hē hit understandan mæg**; **~ andan tō** bear malice | behave to: **yfele ~** ill-treat — *ptc.* **yfele gehæfd** tormented, diseased | have an opinion about: **þā lǣwedan willaþ ~ þone mōnan be þām þe hī hine gesēoþ** | *necessity, obligation w. ger.*: **þone calic þe ic tō drincenne hæbbe** | *to form past tenses w. inflected or (gen.) uninflected ptc.*: **oþ þæt hīe hine ofslægen(ne) hæfdon**.

The parenthesized pronoun in inflected forms such as `(hē) hæfþ, hafaþ` is treated as a subject-pronoun cue identifying a particular paradigm slot. Here it indicates that the following forms represent the third-person singular present, even though Sweet does not state these grammatical properties explicitly. The printed cue is preserved in `<lbl>`, while the grammatical information inferred from it is made presented implicitly in `<gramGrp>`

```xml
<form type="inflected">
   <lbl>(hē)</lbl>
   <gramGrp>
      <gram type="tense" value="present"/>
      <gram type="person" value="third"/>
      <gram type="number" value="singular"/>
   </gramGrp>
   <orth>hæfþ</orth>
   <pc>,</pc>
   <form type="variant">
      <orth>hafaþ</orth>
   </form>
</form>

# **Dealing with the placement of gender, bibl, etc. in cases with complex variants**

`CAS` **hlot**, hlott n. _‘lot,’ part, portion, share_, _Æ, BC, Bl, G_: (+)
_choice, decision_, Æ, AO.  


In cases when a lemma and one or more orthographic variants have only one gender/pos/source/… at the end of all variants, this element applies to the whole entry, so here `n.` should be encoded as a child of entry.

**______________________________________________________________________________**

In cases when there is more than one gender/pos/source/…, like with `wealh-moru`, but the lemma’s gender/pos/source/… is not specified, the first gender/pos/source/ that appears (`f.` in this case)  applies to both lemma and all the variants that precede this gender/pos/source and should, in theory, be encoded as child of entry. However, to preserve the original order of elements in the entry, `f.` is encoded as a cild of lemma's form along with the variants; while any other gender/pos/source (`m.` in this case)  apply to the single variant they follow (to the second one here):

```xml
<entry xml:id="CAS2.wealh-moru" type="mainEntry" xml:lang="ang">
            <form type="lemma">
               <orth>
                  <seg>wealh</seg>
                  <seg>-moru,</seg>
               </orth>
               <form type="variant">
                  <orth extent="part" expand="walh-more"><seg>-more</seg> <seg>(a<lbl rend="sup">1</lbl>)</seg></orth>
               </form>
               <gramGrp>
                  <gram type="pos" value="noun"/>
                  <gram type="gender" value="feminine">f.</gram>,
               </gramGrp>
               <form type="variant">
                  <orth extent="part" expand="wealh-mora"><seg>-mora</seg></orth>
                  <gramGrp>
                     <gram type="gender" value="masculine">m.</gram>
                  </gramGrp>
               </form>
            </form>  
            ...
```       

**______________________________________________________________________________**

`CAS` **ǣg-hwā** mf., -hwæt, n. pron. _each one, every one, everything, who or
whatever_. ǣghwǣt neut. _anything_. [ǣg = āgi]

In this case while `n.` only applies to the second variant, the pos `pron.` applies to the whole entry.

**______________________________________________________________________________**

`CAS` **ād-faru** f. ds. -fære _way or path to the funeral pile_, B 3010.

In cases with inflected variant forms like here, the gender that comes after the lemma applies to the whole entry, while the grammatical information that precedes each of the inflected forms applies to them. In this case the headword does not have a comma separating it from the variants.




```xml
<entry xml:id="CAS2.ād-faru" type="mainEntry" xml:lang="ang">
            <form type="lemma">
               <orth type="segmented">
                  <seg>ād</seg>
                  <seg>-faru</seg>
               </orth>
            </form>
            <gramGrp>
               <gram type="pos" value="noun"/>
               <gram type="gender" value="feminine">f.</gram>
            </gramGrp>
            <form type="inflected">
               <gramGrp>
                  <gram type="case" value="dative">d</gram>
                  <gram type="number" value="singular">s.</gram>
               </gramGrp>
               <orth type="segmented" extent="part" expand="ād-fære"><seg>-fære</seg></orth>
            </form>
            ...

# **Dealing with homographs**

`CAS` **āgan** **I.** (conj. Wt 546) _to own, possess, have, obtain_, _An, Bo, Ma,
Mt_; AO, CP: _give, give up, deliver, restore_: _have to pay, ‘owe,’_
_Mt, Lk_: _have to do_, ANS 123·417. ā. ūt _find out, discover_. **II.**
pret. 3 sg. of āginnan (onginnan).


Our approach to encoding entries with roman numerals  `n=I./II./…` was originally to add this information to the first sense tag following a roman number (signifying the start of an homonymic entry). It was later moved from sense to hom. entry;  the roman number itself was originally encoded as `<num>`, but later changed to `<lbl type="homNum">` (according to TEI Lex-0)  and moved from sense to homonymic entry.

```xml
<entry xml:id="CAS2.āgan_1" type="homonymicEntry" xml:lang="ang" n="I">
     <lbl type="homNum">I.</lbl>
```

# **Dealing with `<gramGrp>` values**

Normalized values are assigned to all `<gram>` elements according to the TEI Lex-0 guidelines (except `inflectionType`, or `construction` where nothing is abbreviated)

```xml
<gram type="pos" value="verb"/>
<gram type="tense" value="past">pret.</gram>
<gram type="person" value="third">3</gram>
<gram type="number" value="singular">sg.</gram>

```

# **Dealing with bibliographic references / attestation sources**

Source references are encoded as bibliographic attestations usually attached to the relevant sense, inside citations (sometimes they may appear inside forms and other places). They may consist of a source abbreviation alone, such as `Bo`, or include a more precise location, such as `Bo 144^31`, with page and line numbers encoded separately. The `@source `attribute links the abbreviation to the corresponding bibliographic record in the TEI header.

```xml
<bibl type="attestation" source="#Bo">
     <title>Bo</title>
     <citedRange unit="page">144</citedRange>
     <citedRange unit="line">31</citedRange>
</bibl>

# **Dealing with placement of `<gramGrp>`**

`CAS` **bǣtan** **I.** _to bait, hunt, worry_, Æ. [v. ‘_bait_’]: (BT) _make fast_?
(Sedgef. = III). **II.** _to beat against the wind_? #Bo# 144^31. [_ON._
beita] **III.** _to furnish with a bit or bridle, saddle, curb_.

`<gramGrp>` is in most cases placed directly under the main `<entry>`. One of the exceptions is when this would disrupt the order of information in the printed entry, as in the `wealh-moru` example. In such cases, `<gramGrp>` is placed inside `<form type="lemma">` so that the encoded sequence follows the source.

```xml
<entry xml:id="CAS2.bǣtan" type="mainEntry" xml:lang="ang">
            <form type="lemma">
               <orth>bǣtan</orth>
            </form>
            <gramGrp>
               <gram type="pos" value="verb"/>                                    
            </gramGrp>
            ...

**______________________________________________________________________________**

`<gramGrp>` follows the TEI Lex-0 principle of default inheritance. Grammatical properties declared at a higher level apply to the descendant structures within the `<entry>`, unless a lower-level `<gramGrp>` provides a different value for the same property. Entry-level `<gram>` elements should therefore contain properties that function as defaults for the entry.

A nested homonym or related `<entry>` does not automatically require its own `<gramGrp>`. One is added only when the nested entry has grammatical properties of its own or when it differs from the inherited description. Local properties override the corresponding higher-level properties, while other inherited properties remain applicable.

`CAS` **āwiht** **I.** n. _‘aught,’ anything, something_, _Ps_. **II.** adv. _at all, by any means_. tō āhte _at all_. **III.** _good, of value_, Æ.

```xml
<entry xml:id="CAS2.āwiht_2" type="homonymicEntry" xml:lang="ang" n="II">
               <lbl type="homNum">II.</lbl>
               <gramGrp>
                  <gram type="pos" value="adverb">adv.</gram>
               </gramGrp>
               ...

**______________________________________________________________________________**

`CAS` **+bǣte**, +bǣtel n. _bit, bridle_, pl. _trappings_, BH.

Here the label `n.` applies to the entry as a whole and is therefore encoded in an entry-level `<gramGrp>`, from which it is inherited by both senses. The label pl. applies only to the meaning trappings and is encoded in a separate `<gramGrp>` within the second `<sense>`.

```xml
<entry xml:id="CAS2.bǣte" type="mainEntry" xml:lang="ang">
            <form type="lemma">
               ...
            </form>                    
            <form type="variant">
               ...
            </form>
            <gramGrp>
               <gram type="pos" value="noun"/>
               <gram type="gender" value="neuter">n.</gram>  
            </gramGrp>
            <sense xml:id="CAS2.bǣte.1">
              <sense xml:id="CAS2.bǣte.1.1">                                           
               ...
              </sense>
              <sense xml:id="CAS2.bǣte.1.2">                                                       
                  <gramGrp>
                     <gram type="number" value="plural">pl.</gram>
                  </gramGrp>
                  <cit type="translationEquivalent" xml:lang="en">
                    ...

Also, usually in this case `... bridle pl. trappings, ...` there should be a sense separating metamark `:` or `;`  after bridle and before `pl.` (or often `in pl.`), but this time it’s just a whitespace. Anyway it is treated as `;` and a separate `<sense>` is added for the sense in plural.

# **Dealing with XR**


`CAS` **wāt** **I.** pres. 3 sg. of witan. **II.** pret. 3 sg. of wītan. **III.** = wǣt

Cross-references are encoded with `<xr type="related">`. They are usually introduced in the printed entries by markers such as `=`, `==`, `see`, or `cp.`. Inflected-form entries instead typically provide grammatical information followed by of and the referenced headword; the grammatical information is encoded separately in `<gramGrp>`, while `of` is preserved in `<lbl>` and the referenced headword is encoded in `<ref>`.

In these and similar cases, `@expand="orthographic variant"` or `@expand="inflected form"` is added to `<xr type="related">` to make the type of relationship explicit. The entry for `wāt` contains both types:

```xml
...
<gramGrp>
     <gram type="pos" value="verb"/>
     <gram type="tense" value="present">pres.</gram>
     <gram type="person" value="third">3</gram>
     <gram type="number" value="singular">sg.</gram>
</gramGrp>
<xr type="related" expand="inflected form">
     <lbl>of</lbl>
     <ref target="#witan" type="entry">witan</ref>
</xr>
...
```
And
```xml
...
<xr type="related" expand="orthographic variant">
     <lbl>=</lbl>
     <ref target="#wǣt" type="entry">wǣt</ref>
</xr>
...


**______________________________________________________________________________**

`CAS` **bæc I.** n. ‘_back_,’ _Bo, MtL, Ps_; CP. on b. _backwards, behind,
‘aback.’_ on b. settan, lǣtan _to neglect_. ofer b. _backwards, back_.
under b. _behind, backwards, back_. ofer b. būgan _to turn back, flee_.
clǣne b. habban _to be straightforward, honest_, LL 128[5]. **II.** mfn.
_beck, brook_. v. ES 29·411.

Sometimes sources are presented as a reference as in `v. ES 29·411`. In these cases the encoding is:


```xml
<xr type="related" expand="bibliographic reference"><lbl norm="vide">v.</lbl><ref target="#ES" type="bibl">ES 29·411</ref></xr>

# **Dealing with implicit POS**

`<gram type="pos" value="..."/>` has been added to every entry that doesn't explicitly mention the pos when it can be inferred from other markers such as gender or verb inflection type.

# **Dealing with `+`, `±` (*ge-* prefix)**

According to the dictionary’s front matter,
`+` = `ge-`.

`±` indicates that the Ags. word to which it is prefixed is found both with and without the prefix `ge-`.


`CAS` **±edcwician** (cwyc-, cuc-) _to re-quicken, revive_, Æ.

For `±` attached to the word like in `±edcwician`, the encoding is:

```xml
...
<form type="lemma">
     <orth extent="full" expand="edcwician, ge-edcwician">
     <lbl expand="ge-_optional">±</lbl>
     <seg>edcwician</seg>
     </orth>
</form>
...

For `+` attached to the word like in `+bǣte`, the encoding is:

```xml
...
<form type="lemma">
     <orth extent="full" expand="ge-bǣte">
     <lbl expand="ge-">+</lbl>
     <seg>bǣte</seg>
     </orth>
</form>
...



**______________________________________________________________________________**

In cases when the lemma with `ge-` prefix is followed by variant segments like in `±edcwician (cwyc-, cuc-)`, the prefix that the lemma has should be inherited by both variants as well. Both expanded forms are listed, `<seg>` is used since both variants are segments of the full form.


```xml
...
<form type="variant">
     <pc>(</pc>
     <orth extent="part" expand="edcwycian, ge-edcwycian">
     <seg>cwyc-</seg>
     </orth>
</form>
<form type="variant">
     <orth extent="part" expand="edcucian, ge-edcucian">
     <seg>cuc-</seg>)
     </orth>
     <pc>)</pc>     
</form>
...


**______________________________________________________________________________**

`CAS` **±etan**^5 (ea, eo) _to ‘eat,’_ _Æ, AO, Jn_: _devour, consume_, _Jn_; CP:
(reflex.) _provision oneself_: (+) _eat together_.

`CAS` **līhtan** (ē, ēo, ȳ) **I.** (±) _to make _‘light_,’ easy, relieve, alleviate_,
_CP, Lcd, LL_: _dismount, ‘alight,’_ _BH_ (y). **II.** (ie, ȳ) _to lighten,
illuminate, give light, shine_, _Æ, Jn_ (ȳ): _grow light, dawn_, _Da_
(ȳ): ‘_light,’ kindle_.

For cases when `+` or  `±` appears on its own, usually in parentheses before senses: `... (+) eat together.` and `... (±) to make ‘light,’ ...`, we suggest an approach as encoded in the following examples:

```xml
...
<sense xml:id="CAS2.etan.1.4">
                  <form type="variant">
                     <orth extent="prefix" expand="ge-etan"><lbl expand="ge-">(+)</lbl>
                     </orth>
                  </form>
                  <cit type="translationEquivalent" xml:lang="en">
                    ...
```

```xml
<ptr target="#CAS2.lyhtan_var"/>

# **Dealing with run-on entries**

`SWEET` **hēah|fȳr** † *n.* high flame.

**~gāst** † *m.* lofty spirit.

Sweet’s run-on entries follow the main entry as shorter subordinate entries. Their headword is abbreviated with `~`, which replaces the element shared with the preceding headword where it is split by `|`, and is usually followed by its own grammatical label and translation equivalent.

```xml
<entry>
  ...
   <entry xml:id="SWT.hēahgāst" type="relatedEntry" xml:lang="ang">
      <form type="lemma">
         <orth extent="part" expand="hēahgāst">
            <seg>~gāst</seg>
         </orth>
      </form>
      <usg type="textType" value="attested in poetic texts">†</usg>
      <gramGrp>
         <gram type="pos" value="noun"/>
         <gram type="gen" value="masculine">m.</gram>
      </gramGrp>
      <sense xml:id="SWT.hēahgāst.1">
        <sense xml:id="SWT.hēahgāst.1.1">
          <cit type="translationEquivalent" xml:lang="en">
             <quote>lofty spirit</quote>
          </cit>
        </sense>
      </sense>
      <metamark>.</metamark>
   </entry>
   ...
</entry>

# **Dealing with examples**

`SWEET` **gehag|ian** *impers.* — **gif him** (*or* **hine**) **tō þām ~ige** if he has the money, can afford it.

Examples occur only in *Sweet*. They are introduced by either `:` or `—` and are encoded as `<cit type="example">`. The Old English example is placed in `<quote>`, while any accompanying English translation is encoded within the same example citation as a nested `<cit>`.
Examples usually follow a translation equivalent within a `<sense>`. In some entries, however, the example constitutes the entire content of the sense, as in `gehagian`. The `<sense>` is still retained and contains only the example citation:

```xml
...
<sense xml:id="SWT.gehagian.1">
   <sense xml:id="SWT.gehagian.1.1">
      <gramGrp>
         <gram type="valency" value="impersonal">impers.</gram>
      </gramGrp>
      <pc>—</pc>
      <cit type="example" xml:lang="ang">
         <quote>gif him (or hine) tō þām <ref type="form" scope="currentEntry" value="gehag">~</ref>ige</quote>
         <cit type="translation" xml:lang="en">
            <quote>if he has the money, can afford it</quote>
         </cit>
      </cit>
   </sense>
</sense>
...

# **Dealing with related entreis**

`CAS` **bæc I.** n. ‘_back_,’ _Bo, MtL, Ps_; CP. on b. _backwards, behind,
‘aback.’_ on b. settan, lǣtan _to neglect_. ofer b. _backwards, back_.
under b. _behind, backwards, back_. ofer b. būgan _to turn back, flee_.
clǣne b. habban _to be straightforward, honest_, LL 128[5]. **II.** mfn.
_beck, brook_. v. ES 29·411.

In *CAS*, related entries are additional lemmas or multiword expressions that follow a full stop (`.`) within the entry. They consistently occur as the final major structural element of an entry or a homonymic entry, making their interpretation as related entries relatively secure. Each is assigned its own unique `@xml:id` and encoded as a related entry with a `<form type="lemma">`. Where the related entry is a multiword expression, this is additionally indicated by `<gram type="mwe">`

```xml
<entry xml:id="CAS2.clæne_bæc_habban" type="relatedEntry" xml:lang="ang">
  <form type="lemma">
    <orth>clǣne <ref type="form" scope="currentEntry" value="bæc">b.</ref> habban</orth>
  </form>
  <gramGrp>
    <gram type="mwe" value="fixed_combination"/>
  </gramGrp>
  ...

**______________________________________________________________________________**

`SWEET` **cild** *n., pl.* **cild(ru)** child — **wearþ mid (þām) ~e** became pregnant. **~a hierde** schoolmaster, paedagogus. **~a mæssedæg** Innocents’-day. **~trog** cradle.

For *Sweet*, related entries are identified more cautiously than in *CAS*. Currently, a form following a full stop (`.`) and accompanied by its own translation is treated as a related entry. Such forms often use `~` to replace the shared element of the main headword. As the boundary between subordinate lexical items and illustrative material (i.e., related entries and examples) is not always clear in *Sweet*, this part of the policy may require further refinement. In this case, `wearþ mid (þām) cilde` is considered an example, while `cilda hierde`, `cilda mæssedæg` and `cildtrog` are encoded as related entries

```xml
<entry xml:id="SWT.cilda_hierde" type="relatedEntry" xml:lang="ang">
   <form type="lemma">
      <orth><ref type="form" scope="currentEntry" value="cild">~</ref>a hierde</orth>
   </form>
   <gramGrp>
      <gram type="mwe" value="fixed_combination"/>
   </gramGrp>
   ...

# **Dealing with strong/weak**

## Verbs

`CAS` **±etan**^5 (ea, eo) _to ‘eat,’_ _Æ, AO, Jn_: _devour, consume_, _Jn_; CP:
(reflex.) _provision oneself_: (+) _eat together_.

`SWEET` **wrēon** 6, 7 *prt.* **wrāh**, **ēa**, *pl.* **wrigon**, **u**, *ptc.* **wrigen**, **o** cover *rare in W*.

For verbs, the distinction between strong and weak classes is represented through the pos value, while the strong verb inflectional class (here `^5`) is encoded separately with `<gram type="inflectionType">`. Where the number is printed as a superscript, this formatting is retained with `<hi>`
```xml
<gramGrp>
   <gram type="pos" value="strong verb"/>
   <gram type="inflectionType"><hi rend="superscript">5</hi></gram>          
</gramGrp>


When two alternative strong verb classes are given as in `wrēon 6, 7`, both are preserved as a separate `<gram>` elements:
```xml
<gramGrp>
   <gram type="pos" value="strong verb"/>
   <gram type="inflectionType">6</gram>
   <pc>,</pc>
   <gram type="inflectionType">7</gram>
</gramGrp>

## Adjectives, nouns

`CAS` **gearo** (gearu) **I.** (wk gearwa) gsmn. -(o)wes; asm. gearone; napn. gearu _prepared, ready, equipped, finished_, #Chr#, W. [‘_yare_’] **II.** = gearwe
adv.

For adjectives and nouns, strong and weak inflection is encoded as declension with `<gram type="declension">`. This information may apply to the entry as a whole or only to a particular form. In the *CAS* entry for `gearo`, in `(wk gearwa)`, the label `wk` applies specifically to the variant `gearwa`. The entry-level grammatical information identifies the word as an adjective, while the weak declension is placed inside the corresponding variant form
```xml
<entry>
  ...
  <gramGrp>
    <gram type="pos" value="adjective"/>
  </gramGrp>
  ...
  <form type="variant">
    <gramGrp>
        <gram type="declension" value="weak">wk</gram>
    </gramGrp>
    <orth>gearwa</orth>
  </form>
  ...

**______________________________________________________________________________**

`CAS` **piða** wm. _inward part, ‘pith,’_ _Bo_; CP.

The same treatment is used for nouns. In an abbreviation such as `wm.`, the first letter indicates weak declension and the second - masculine gender. These are encoded as separate grammatical properties:
```xml
<gramGrp>
  <gram type="pos" value="noun"/>
  <gram type="declension" value="weak">w</gram>
  <gram type="gender" value="masculine">m.</gram>
</gramGrp>

# **Dealing with punctuation**

Punctuation that contributes to the structural organization of an entry is encoded with `<metamark>`. In *CAS*, this includes `:` and (rarely) `;` used to separate senses, as well as full stops (`.`) that mark the end of an entry or a related entry. Where the punctuation functions as a sense boundary, it is encoded with an additional `@function="senseSeparator"`.

In *Sweet*, `;` and `|` or `||` are treated as metamarks. Full stops that delimit the end of an entry or related entry are also preserved as `<metamark>`. Punctuation introducing examples, particularly `:` and `—`, is treated in the same way. At present, only the function `senseSeparator` is assigned explicitly; other structural punctuation is encoded as `<metamark>` without a more specific `@function`.

Punctuation that does not perform a clear structural role is either retained within the surrounding textual string or represented separately with `<pc>`. This distinction, however, is not yet applied fully consistently across the encodings. The treatment of punctuation therefore remains an area requiring further normalization.
